```python
# Business Question 1
# What is the distribution of Movies vs TV Shows on Netflix?
content_type = netflix["type"].value_counts()

content_type
```

```python
plt.figure(figsize=(8,5))

content_type.plot(
    kind="bar",
    edgecolor="black"
)

plt.title("Distribution of Movies and TV Shows on Netflix", fontsize=14)
plt.xlabel("Content Type")
plt.ylabel("Number of Titles")

plt.show()
```

```python
# Netflix offers significantly more Movies than TV Shows, indicating a stronger focus on movie content within its catalog.
# Netflix can continue investing in Movies while monitoring the growing popularity of TV Shows to maintain a balanced content library.
```

```python
# Business Question 2
# What percentage of Netflix content is Movies vs TV Shows?
content_percentage = (
    netflix["type"]
    .value_counts(normalize=True)
    * 100
).round(2)

content_percentage
```

```python
plt.figure(figsize=(7,7))

plt.pie(
    content_percentage,
    labels=content_percentage.index,
    autopct="%1.1f%%",
    startangle=90
)

plt.title("Percentage of Movies and TV Shows")

plt.show()
```

```python
# Business Insight

# Approximately 69.9% of Netflix's catalog consists of Movies, while 30.4% consists of TV Shows.
```

```python
# Make a working copy
netflix = df.copy()

# Remove duplicates
netflix.drop_duplicates(inplace=True)

# Convert date column
netflix["date_added"] = pd.to_datetime(netflix["date_added"], errors="coerce")

# Create new features
netflix["year_added"] = netflix["date_added"].dt.year
netflix["month_added"] = netflix["date_added"].dt.month_name()
netflix["day_added"] = netflix["date_added"].dt.day_name()

# Fill missing values
netflix["director"] = netflix["director"].fillna("Unknown")
netflix["cast"] = netflix["cast"].fillna("Unknown")
netflix["country"] = netflix["country"].fillna("Unknown")
netflix["rating"] = netflix["rating"].fillna("Not Rated")

# Remove extra spaces
text_columns = ["title", "director", "cast", "country", "rating", "listed_in"]

for col in text_columns:
    netflix[col] = netflix[col].str.strip()

# Split duration
netflix[["duration_value", "duration_unit"]] = netflix["duration"].str.split(" ", n=1, expand=True)
netflix["duration_value"] = pd.to_numeric(netflix["duration_value"], errors="coerce")
```

```python
# Business Question 3
# Which countries produce the most Netflix content?
country_df = netflix.copy()

country_df["country"] = country_df["country"].str.split(",")

country_df = country_df.explode("country")

country_df["country"] = country_df["country"].str.strip()

top_countries = country_df["country"].value_counts().head(10)

top_countries
```

```python
plt.figure(figsize=(10,6))

top_countries.sort_values().plot(kind="barh", edgecolor="black")

plt.title("Top 10 Countries Producing Netflix Content")
plt.xlabel("Number of Titles")
plt.ylabel("Country")

plt.show()
```

```python
# Business Insight:

# The United States contributes the highest number of titles, followed by India and the United Kingdom.

# Recommendation:

# Netflix can strengthen partnerships in emerging markets where production is increasing.
```

```python
# Business Question 4
# Which genres are most popular? 
genre_df = netflix.copy()

genre_df["listed_in"] = genre_df["listed_in"].str.split(",")

genre_df = genre_df.explode("listed_in")

genre_df["listed_in"] = genre_df["listed_in"].str.strip()
top_genres = genre_df["listed_in"].value_counts().head(10)

top_genres
```

```python
plt.figure(figsize=(10,6))

top_genres.sort_values().plot(
    kind="barh",
    edgecolor="black"
)

plt.title("Top 10 Genres on Netflix")

plt.xlabel("Number of Titles")

plt.show()
```

```python
# Business Insight

# International Movies and Dramas dominate Netflix's catalog.

# Recommendation

# Netflix should continue investing in internationally popular genres while exploring underrepresented categories.
```

```python
# Business Question 5
# How has Netflix's content library grown over time? 
content_growth = (
    netflix["year_added"]
    .value_counts()
    .sort_index()
)

content_growth
plt.figure(figsize=(12,5))

content_growth.plot(
    marker="o",
    linewidth=2
)

plt.title("Content Added to Netflix by Year")
plt.xlabel("Year")
plt.ylabel("Number of Titles")

plt.grid(True)

plt.show()
```

```python
# Business Insight

# Netflix experienced rapid catalog growth between 2016 and 2019 before stabilizing.

# Recommendation

# Analyze whether newer content additions align with subscriber demand across regions.
```

```python
# Business Question 6
# Which ratings are most common? 
ratings = netflix["rating"].value_counts()

ratings
plt.figure(figsize=(10,5))

ratings.plot(
    kind="bar",
    edgecolor="black"
)

plt.title("Distribution of Content Ratings")
plt.xlabel("Rating")
plt.ylabel("Number of Titles")

plt.show()
```

```python
# Business Insight

# # TV-MA and TV-14 are among the most common ratings, indicating a strong focus on mature and teen audiences.
```

```python
# Business Question 7
# Who are the Top 10 Directors? 
director_df = netflix.copy()

director_df["director"] = director_df["director"].str.split(",")

director_df = director_df.explode("director")

director_df["director"] = director_df["director"].str.strip()

top_directors = director_df[
    director_df["director"] != "Unknown"
]["director"].value_counts().head(10)

top_directors
plt.figure(figsize=(10,6))

top_directors.sort_values().plot(
    kind="barh",
    edgecolor="black"
)

plt.title("Top 10 Directors")

plt.show()
```

```python
# Business Question 8
# Movie Duration Distribution
movies = netflix[
    netflix["type"]=="Movie"
]
plt.figure(figsize=(10,5))

movies["duration_value"].plot(
    kind="hist",
    bins=30,
    edgecolor="black"
)

plt.title("Movie Duration Distribution")

plt.xlabel("Minutes")

plt.show()
```
